```python
import tensorflow as tf
import numpy as np

print("TensorFlow version:", tf.__version__)

# Basic tensor structures
num_scalar = tf.constant(7)
num_vector = tf.constant([10, 20, 30])
num_matrix = tf.constant([[1.0, 2.0], [3.0, 4.0]])
num_tensor = tf.constant([[[1], [2]], [[3], [4]]])

print("Shapes:")
print(num_scalar.shape, num_vector.shape, num_matrix.shape, num_tensor.shape)

# Data types and casting
float_values = tf.constant([1.7, 7.4])
int_values = tf.constant([7, 10])

float16_values = tf.cast(float_values, tf.float16)
float32_values = tf.cast(int_values, tf.float32)

# Indexing and reshaping
sample_matrix = tf.constant([[10, 7], [3, 4]])

column_slice = sample_matrix[:, -1]
added_axis = tf.expand_dims(sample_matrix, axis=-1)
restored_shape = tf.squeeze(added_axis, axis=-1)

# Arithmetic operations (broadcasting)
matrix_a = tf.constant([[10.0, 7.0], [3.0, 4.0]])

print("Add 10:\n", matrix_a + 10)
print("Multiply by 2:\n", matrix_a * 2)

# Matrix multiplication example
matrix_x = tf.constant([[1, 2], [3, 4], [5, 6]], dtype=tf.float32)
matrix_y = tf.constant([[7, 8], [9, 10], [11, 12]], dtype=tf.float32)

result_matrix = tf.matmul(matrix_x, tf.transpose(matrix_y))

# Aggregation operations
random_values = tf.constant(np.random.randint(0, 100, size=10))

print(
    "Min:", tf.reduce_min(random_values).numpy(),
    "Max:", tf.reduce_max(random_values).numpy(),
    "Mean:", tf.reduce_mean(tf.cast(random_values, tf.float32)).numpy(),
    "Sum:", tf.reduce_sum(random_values).numpy()
)

# One-hot encoding example
indices = tf.constant([0, 1, 2, 3])
one_hot_encoded = tf.one_hot(indices, depth=4)

# NumPy and TensorFlow conversion
numpy_array = np.array([3.0, 7.0, 10.0])

tensor_from_np = tf.constant(numpy_array)
numpy_back = tensor_from_np.numpy()

print("All operations completed successfully")
```
